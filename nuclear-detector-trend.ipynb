```python
import matplotlib.pyplot as plt

# Given data
voltages = [20, 40, 60, 100, 120]  # Voltage values in V
counts = [10, 14, 17, 25, 18]  # Corresponding counts

# Plot the data
plt.figure(figsize=(8, 5))
plt.plot(voltages, counts, marker='o', linestyle='-', color='b', label="Counts vs Voltage")

# Labels and title
plt.xlabel("Voltage (V)")
plt.ylabel("Number of Counts")
plt.title("Number of Counts vs Voltage")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
voltages = np.array([0,100,200,250,270,290,300,320,340,360,380,400,420,440,460,480,500,520,540,560,580,600])  # Voltage values in V
counts = np.array([0,0,0,0,0,0,1731,16219,20045,21641,24433,26096,27306,28239,29110,29384,30192,30636,31311,31589,32113,32433])  # Corresponding counts

# Fit a linear trendline (polynomial of degree 1)
coefficients = np.polyfit(voltages, counts, 1)
trendline = np.poly1d(coefficients)

# Generate trendline values
voltage_range = np.linspace(min(voltages), max(voltages), 100)
trend_counts = trendline(voltage_range)

# Calculate R-squared value
y_pred = trendline(voltages)  # Predicted values from the trendline
ss_res = np.sum((counts - y_pred) ** 2)  # Residual sum of squares
ss_tot = np.sum((counts - np.mean(counts)) ** 2)  # Total sum of squares
r_squared = 1 - (ss_res / ss_tot)  # R-squared calculation

# Plot the data with a connecting line between points and the trendline
plt.figure(figsize=(8, 5))
plt.plot(voltages, counts, 'bo-', label="Data Points")  # Line connecting data points
plt.plot(voltage_range, trend_counts, 'r-', 
         label=f"Trendline: y = {coefficients[0]:.2f}x + {coefficients[1]:.2f}\nR² = {r_squared:.4f}")

# Labels and title
plt.xlabel("Voltage (V)")
plt.ylabel("Number of Counts")
plt.title("Number of Counts vs Voltage with Trendline and R-squared")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()

# Display R-squared value
print(coefficients)
print(r_squared)
```
